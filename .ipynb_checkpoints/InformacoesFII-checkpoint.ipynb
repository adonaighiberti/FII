```python
#Importa Bibliotecas

from bs4 import BeautifulSoup
import requests
import time
```

```python
#Chama página com lista de FIIS

url_listafiis = "https://fiis.com.br/lista-de-fundos-imobiliarios/"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.121 Safari/537.36"
}

response = requests.get(url_listafiis, headers=headers)

if response.status_code == 200:
    
    fiis = BeautifulSoup(response.text, 'html.parser')

else:
    print(f"Erro ao acessar a página: {response.status_code}")
```

```python
# Trazer todos os Fiis em uma lista
recorta_nome_fiis = fiis.find_all("div", class_="tickerBox__title", attrs={"data-element": "ticker-box-title"})

nome_fiis = []
for fii in recorta_nome_fiis:
    nome_fiis.append(fii.text)
```

```python
#Gera links para cada FII

url_fiis = "https://fiis.com.br/"
url_fiis_completa = []

for i in nome_fiis:

    url_fiis_temp = url_fiis + i
    url_fiis_completa.append(url_fiis_temp)
```

```python
#Traz informação de cada FII

Titulo = []
Nomefii = []
Administrador = []
dividendo_yeld_ano = []
ultimo_dividendo = []
CNPJ = []
Nome_Pregao = []
QTD_Cotas = []
Patrimonio = []
Segmento = []
Publico_Alvo = []
data_base = []
pg_div_dt = []
cotacao_base = []
dividend_yeld = []
cotacao_atual = []
valor_caixa = [] 
liquidez_dia = []
vrPatr_cota = [] 
qdtCotistas = [] 
partIFIX = []

segundatentativa = []

erros = []

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.121 Safari/537.36"
}

for j in url_fiis_completa:

    retorno = requests.get(j, headers=headers, timeout=(5, 10))
    
    if retorno.status_code == 200:
    
        fii = BeautifulSoup(retorno.text, 'html.parser')
        indicadores = fii.find_all("div", class_="headerTicker__content__info")
        Dados = fii.find_all("div", class_="moreInfo wrapper")
        tabeladiv = fii.find_all("div", class_="yieldChart__table__bloco yieldChart__table__bloco--rendimento")
        detalhes_cota = fii.find("div", class_="quotations-details")
        indicadores_caixa = fii.find("div", class_="wrapper indicators")
        
        Titulo.append(fii.title.string)
        Nomefii.append(fii.find("h1").string)
        Administrador.append(fii.find("p", class_="informations__adm__name").text)
        dividendo_yeld_ano.append(indicadores[0].find(string="Dividend Yield").find_previous("b").text.strip())
        ultimo_dividendo.append(indicadores[0].find(string="Último Rendimento").find_previous("b").text.strip())
        CNPJ.append(Dados[0].find("span", string="CNPJ").find_next("b").text.strip())
        Nome_Pregao.append(Dados[0].find("span", string="Nome no Pregão").find_next("b").text.strip())
        QTD_Cotas.append(Dados[0].find("span", string="Número de Cotas").find_next("b").text.strip())
        Patrimonio.append(Dados[0].find("span", string="Patrimônio").find_next("b").text.strip())
        Segmento.append(Dados[0].find("span", string="Segmento").find_next("b").text.strip())
        Publico_Alvo.append(Dados[0].find("span", string="Publico Alvo").find_next("b").text.strip())
        cotacao_atual.append(detalhes_cota.find("span", class_="value").text.strip())
        valor_caixa.append(indicadores_caixa.find(string="Valor em caixa").find_previous("b").text.strip())
        liquidez_dia.append(indicadores_caixa.find(string="Liquidez média diária").find_previous("b").text.strip())
        vrPatr_cota.append(indicadores_caixa.find(string="Val. Patrimonial p/Cota").find_previous("b").text.strip())
        qdtCotistas.append(indicadores_caixa.find(string="N° de Cotistas").find_previous("b").text.strip())
        partIFIX.append(indicadores_caixa.find(string="Participações no IFIX").find_previous("b").text.strip())
        
        try:
        
            data_base.append(tabeladiv[0].find_all("div", class_="table__linha")[1].text.strip())
            pg_div_dt.append(tabeladiv[0].find_all("div", class_="table__linha")[2].text.strip())
            cotacao_base.append(tabeladiv[0].find_all("div", class_="table__linha")[3].text.strip())
            dividend_yeld.append(tabeladiv[0].find_all("div", class_="table__linha")[4].text.strip())
        
        except:
        
            data_base.append("")
            pg_div_dt.append("")
            cotacao_base.append("")
            dividend_yeld.append("")
    
    elif retorno.status_code == 404:
    
        segundatentativa = j.replace("11" , "")
        
        retorno = requests.get(segundatentativa, headers=headers, timeout=(5, 10))
        
        if retorno.status_code == 200:
        
            fii = BeautifulSoup(retorno.text, 'html.parser')
            indicadores = fii.find_all("div", class_="headerTicker__content__info")
            Dados = fii.find_all("div", class_="moreInfo wrapper")
            tabeladiv = fii.find_all("div", class_="yieldChart__table__bloco yieldChart__table__bloco--rendimento")
            detalhes_cota = fii.find("div", class_="quotations-details")
            indicadores_caixa = fii.find("div", class_="wrapper indicators")
            
            Titulo.append(fii.title.string)
            Nomefii.append(fii.find("h1").string)
            Administrador.append(fii.find("p", class_="informations__adm__name").text)
            dividendo_yeld_ano.append(indicadores[0].find(string="Dividend Yield").find_previous("b").text.strip())
            ultimo_dividendo.append(indicadores[0].find(string="Último Rendimento").find_previous("b").text.strip())
            CNPJ.append(Dados[0].find("span", string="CNPJ").find_next("b").text.strip())
            Nome_Pregao.append(Dados[0].find("span", string="Nome no Pregão").find_next("b").text.strip())
            QTD_Cotas.append(Dados[0].find("span", string="Número de Cotas").find_next("b").text.strip())
            Patrimonio.append(Dados[0].find("span", string="Patrimônio").find_next("b").text.strip())
            Segmento.append(Dados[0].find("span", string="Segmento").find_next("b").text.strip())
            Publico_Alvo.append(Dados[0].find("span", string="Publico Alvo").find_next("b").text.strip())
            cotacao_atual.append(detalhes_cota.find("span", class_="value").text.strip())
            valor_caixa.append(indicadores_caixa.find(string="Valor em caixa").find_previous("b").text.strip())
            liquidez_dia.append(indicadores_caixa.find(string="Liquidez média diária").find_previous("b").text.strip())
            vrPatr_cota.append(indicadores_caixa.find(string="Val. Patrimonial p/Cota").find_previous("b").text.strip())
            qdtCotistas.append(indicadores_caixa.find(string="N° de Cotistas").find_previous("b").text.strip())
            partIFIX.append(indicadores_caixa.find(string="Participações no IFIX").find_previous("b").text.strip())
            
            try:
            
                data_base.append(tabeladiv[0].find_all("div", class_="table__linha")[1].text.strip())
                pg_div_dt.append(tabeladiv[0].find_all("div", class_="table__linha")[2].text.strip())
                cotacao_base.append(tabeladiv[0].find_all("div", class_="table__linha")[3].text.strip())
                dividend_yeld.append(tabeladiv[0].find_all("div", class_="table__linha")[4].text.strip())
            
            except:
            
                data_base.append("")
                pg_div_dt.append("")
                cotacao_base.append("")
                dividend_yeld.append("")
            
        else:
            erro = (f"{j} : {retorno}")
            erros.append(erro)
    
    else:
        erro = (f"{j} : {retorno}")
        erros.append(erro)
        
        time.sleep(2)
```
